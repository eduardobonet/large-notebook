```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def do_plot(is_sin=False):
    x = np.linspace(0, 4*np.pi,50)

    if is_sin:
        y = 2 * np.sin(x)
        c = 'b'
    else:
        y = 2 * x
        c = 'r'
        

    figure(figsize=(20, 6), dpi=100)

    plt.plot(x, y, color=c)
```

```python
do_plot(is_sin = False)
```
